```python
import re
VIVADO_LOG = """
----------------------------------------------
 Triggering Inference Tick 1 at time 1695000
----------------------------------------------
  [Time:              1715000] Streamed Pixel: 220 | Poisson RNG:  99 | Generated Raw Spike: 1
  [Time:              1725000] Streamed Pixel: 220 | Poisson RNG:  35 | Generated Raw Spike: 1
  [Time:              1735000] Streamed Pixel: 220 | Poisson RNG:  32 | Generated Raw Spike: 1
  [Time:              1745000] Streamed Pixel: 220 | Poisson RNG: 198 | Generated Raw Spike: 1
  [Time:              1755000] Streamed Pixel: 220 | Poisson RNG:  75 | Generated Raw Spike: 1
  [Time:              1765000] Streamed Pixel: 220 | Poisson RNG:  75 | Generated Raw Spike: 1
  [Time:              1775000] Streamed Pixel: 220 | Poisson RNG: 251 | Generated Raw Spike: 0
  [Time:              1785000] Streamed Pixel: 220 | Poisson RNG: 255 | Generated Raw Spike: 0
  [Time:              1795000] Streamed Pixel: 220 | Poisson RNG: 189 | Generated Raw Spike: 1
  [Time:              1805000] Streamed Pixel: 220 | Poisson RNG:  66 | Generated Raw Spike: 1
  [Time:              1815000] Streamed Pixel: 220 | Poisson RNG:  51 | Generated Raw Spike: 1
  [Time:              1825000] Streamed Pixel: 220 | Poisson RNG: 117 | Generated Raw Spike: 1
  [Time:              1835000] Streamed Pixel: 220 | Poisson RNG:  95 | Generated Raw Spike: 1
  [Time:              1845000] Streamed Pixel: 220 | Poisson RNG: 255 | Generated Raw Spike: 0
  [Time:              1855000] Streamed Pixel: 220 | Poisson RNG:  32 | Generated Raw Spike: 1
  [Time:              1865000] Streamed Pixel: 220 | Poisson RNG: 229 | Generated Raw Spike: 0
  [Time:              1875000] Streamed Pixel:   0 | Poisson RNG:  64 | Generated Raw Spike: 0
  [Time:              1885000] Streamed Pixel:   0 | Poisson RNG: 109 | Generated Raw Spike: 0
  [Time:              1895000] Streamed Pixel:   0 | Poisson RNG: 137 | Generated Raw Spike: 0

>>> TICK 1 COMPLETE <<<
==============================================
 NEURON | POTENTIAL | FIRED?
----------------------------------------------
    0   |     120   |   0
    1   |     133   |   0
    2   |     145   |   0
    3   |     157   |   0
    4   |     169   |   0
    5   |     181   |   0
    6   |     193   |   0
    7   |     205   |   0
    8   |     217   |   0
    9   |     228   |   0
==============================================


----------------------------------------------
 Triggering Inference Tick 2 at time 2085000
----------------------------------------------
  [Time:              2105000] Streamed Pixel: 220 | Poisson RNG:   2 | Generated Raw Spike: 1
  [Time:              2115000] Streamed Pixel: 220 | Poisson RNG:  71 | Generated Raw Spike: 1
  [Time:              2125000] Streamed Pixel: 220 | Poisson RNG:   5 | Generated Raw Spike: 1
  [Time:              2135000] Streamed Pixel: 220 | Poisson RNG:  42 | Generated Raw Spike: 1
  [Time:              2145000] Streamed Pixel: 220 | Poisson RNG:  23 | Generated Raw Spike: 1
  [Time:              2155000] Streamed Pixel: 220 | Poisson RNG: 158 | Generated Raw Spike: 1
  [Time:              2165000] Streamed Pixel: 220 | Poisson RNG:  95 | Generated Raw Spike: 1
  [Time:              2175000] Streamed Pixel: 220 | Poisson RNG: 217 | Generated Raw Spike: 1
  [Time:              2185000] Streamed Pixel: 220 | Poisson RNG: 128 | Generated Raw Spike: 1
  [Time:              2195000] Streamed Pixel: 220 | Poisson RNG: 120 | Generated Raw Spike: 1
  [Time:              2205000] Streamed Pixel: 220 | Poisson RNG: 105 | Generated Raw Spike: 1
  [Time:              2215000] Streamed Pixel: 220 | Poisson RNG: 131 | Generated Raw Spike: 1
  [Time:              2225000] Streamed Pixel: 220 | Poisson RNG:  10 | Generated Raw Spike: 1
  [Time:              2235000] Streamed Pixel: 220 | Poisson RNG: 115 | Generated Raw Spike: 1
  [Time:              2245000] Streamed Pixel: 220 | Poisson RNG: 103 | Generated Raw Spike: 1
  [Time:              2255000] Streamed Pixel: 220 | Poisson RNG: 210 | Generated Raw Spike: 1
  [Time:              2265000] Streamed Pixel:   0 | Poisson RNG: 130 | Generated Raw Spike: 0
  [Time:              2275000] Streamed Pixel:   0 | Poisson RNG: 156 | Generated Raw Spike: 0
  [Time:              2285000] Streamed Pixel:   0 | Poisson RNG:  11 | Generated Raw Spike: 0

>>> TICK 2 COMPLETE <<<
==============================================
 NEURON | POTENTIAL | FIRED?
----------------------------------------------
    0   |     266   |   0
    1   |     283   |   0
    2   |     308   |   0
    3   |     334   |   0
    4   |     359   |   0
    5   |     385   |   0
    6   |     410   |   0
    7   |     436   |   0
    8   |     461   |   0
    9   |     485   |   0
==============================================


----------------------------------------------
 Triggering Inference Tick 3 at time 2475000
----------------------------------------------
  [Time:              2495000] Streamed Pixel: 220 | Poisson RNG: 109 | Generated Raw Spike: 1
  [Time:              2505000] Streamed Pixel: 220 | Poisson RNG:  74 | Generated Raw Spike: 1
  [Time:              2515000] Streamed Pixel: 220 | Poisson RNG: 173 | Generated Raw Spike: 1
  [Time:              2525000] Streamed Pixel: 220 | Poisson RNG: 254 | Generated Raw Spike: 0
  [Time:              2535000] Streamed Pixel: 220 | Poisson RNG: 144 | Generated Raw Spike: 1
  [Time:              2545000] Streamed Pixel: 220 | Poisson RNG: 102 | Generated Raw Spike: 1
  [Time:              2555000] Streamed Pixel: 220 | Poisson RNG: 193 | Generated Raw Spike: 1
  [Time:              2565000] Streamed Pixel: 220 | Poisson RNG: 225 | Generated Raw Spike: 0
  [Time:              2575000] Streamed Pixel: 220 | Poisson RNG: 230 | Generated Raw Spike: 0
  [Time:              2585000] Streamed Pixel: 220 | Poisson RNG: 191 | Generated Raw Spike: 1
  [Time:              2595000] Streamed Pixel: 220 | Poisson RNG: 140 | Generated Raw Spike: 1
  [Time:              2605000] Streamed Pixel: 220 | Poisson RNG: 157 | Generated Raw Spike: 1
  [Time:              2615000] Streamed Pixel: 220 | Poisson RNG: 159 | Generated Raw Spike: 1
  [Time:              2625000] Streamed Pixel: 220 | Poisson RNG:  48 | Generated Raw Spike: 1
  [Time:              2635000] Streamed Pixel: 220 | Poisson RNG: 232 | Generated Raw Spike: 0
  [Time:              2645000] Streamed Pixel: 220 | Poisson RNG: 228 | Generated Raw Spike: 0
  [Time:              2655000] Streamed Pixel:   0 | Poisson RNG:  26 | Generated Raw Spike: 0
  [Time:              2665000] Streamed Pixel:   0 | Poisson RNG: 220 | Generated Raw Spike: 0
  [Time:              2675000] Streamed Pixel:   0 | Poisson RNG:  40 | Generated Raw Spike: 0

>>> TICK 3 COMPLETE <<<
==============================================
 NEURON | POTENTIAL | FIRED?
----------------------------------------------
    0   |     343   |   0
    1   |     369   |   0
    2   |     402   |   0
    3   |     436   |   0
    4   |     469   |   0
    5   |     502   |   0
    6   |     535   |   0
    7   |     569   |   0
    8   |     602   |   0
    9   |     634   |   0
==============================================


----------------------------------------------
 Triggering Inference Tick 4 at time 2865000
----------------------------------------------
  [Time:              2885000] Streamed Pixel: 220 | Poisson RNG:   3 | Generated Raw Spike: 1
  [Time:              2895000] Streamed Pixel: 220 | Poisson RNG: 173 | Generated Raw Spike: 1
  [Time:              2905000] Streamed Pixel: 220 | Poisson RNG:  46 | Generated Raw Spike: 1
  [Time:              2915000] Streamed Pixel: 220 | Poisson RNG:  18 | Generated Raw Spike: 1
  [Time:              2925000] Streamed Pixel: 220 | Poisson RNG:  88 | Generated Raw Spike: 1
  [Time:              2935000] Streamed Pixel: 220 | Poisson RNG:  87 | Generated Raw Spike: 1
  [Time:              2945000] Streamed Pixel: 220 | Poisson RNG: 214 | Generated Raw Spike: 1
  [Time:              2955000] Streamed Pixel: 220 | Poisson RNG: 123 | Generated Raw Spike: 1
  [Time:              2965000] Streamed Pixel: 220 | Poisson RNG:  77 | Generated Raw Spike: 1
  [Time:              2975000] Streamed Pixel: 220 | Poisson RNG: 218 | Generated Raw Spike: 1
  [Time:              2985000] Streamed Pixel: 220 | Poisson RNG:  32 | Generated Raw Spike: 1
  [Time:              2995000] Streamed Pixel: 220 | Poisson RNG: 198 | Generated Raw Spike: 1
  [Time:              3005000] Streamed Pixel: 220 | Poisson RNG: 238 | Generated Raw Spike: 0
  [Time:              3015000] Streamed Pixel: 220 | Poisson RNG: 242 | Generated Raw Spike: 0
  [Time:              3025000] Streamed Pixel: 220 | Poisson RNG:  79 | Generated Raw Spike: 1
  [Time:              3035000] Streamed Pixel: 220 | Poisson RNG: 226 | Generated Raw Spike: 0
  [Time:              3045000] Streamed Pixel:   0 | Poisson RNG:  50 | Generated Raw Spike: 0
  [Time:              3055000] Streamed Pixel:   0 | Poisson RNG: 225 | Generated Raw Spike: 0
  [Time:              3065000] Streamed Pixel:   0 | Poisson RNG:  47 | Generated Raw Spike: 0

>>> TICK 4 COMPLETE <<<
==============================================
 NEURON | POTENTIAL | FIRED?
----------------------------------------------
    0   |     431   |   0
    1   |     467   |   0
    2   |     509   |   0
    3   |     552   |   0
    4   |     594   |   0
    5   |     636   |   0
    6   |     678   |   0
    7   |     720   |   0
    8   |     762   |   0
    9   |     802   |   0
==============================================


==============================================
--- Simulation Complete ---
==============================================
"""

def verify_snn_hardware(log_text):
    num_inputs = 16
    num_neurons = 10
    
    # Weights programmed as 10 + neuron_index
    weights = [10 + i for i in range(num_neurons)]
    
    # Golden model potentials
    expected_potentials = [0] * num_neurons
    
    # Split log into individual tick sections
    ticks = log_text.split("Triggering Inference Tick")[1:]
    
    for tick_data in ticks:
        tick_num_match = re.search(r'(\d+)', tick_data)
        if not tick_num_match:
            continue
        tick_num = int(tick_num_match.group(1))
        
        # Extract first 16 generated spikes
        spike_matches = re.findall(r'Generated Raw Spike:\s*(\d)', tick_data)
        valid_spikes = [int(s) for s in spike_matches[:num_inputs]]
        num_spikes = sum(valid_spikes)
        
        # Extract hardware scoreboard values
        scoreboard_matches = re.findall(
            r'^\s*(\d+)\s+\|\s+(\d+)\s+\|\s+(\d+)',
            tick_data,
            re.MULTILINE
        )
        actual_potentials = {int(m[0]): int(m[1]) for m in scoreboard_matches}
        
        print(f"\n==============================================")
        print(f" TICK {tick_num} VERIFICATION")
        print(f" Valid Spikes Detected: {num_spikes} out of 16")
        print(f"----------------------------------------------")
        print(f" NEURON | EXPECTED | HARDWARE")
        print(f"----------------------------------------------")
        
        # Golden model update + comparison
        for n in range(num_neurons):
            added_charge = num_spikes * weights[n]
            leak = expected_potentials[n] >> 3  # divide by 8
            expected_potentials[n] = expected_potentials[n] - leak + added_charge
            
            actual = actual_potentials.get(n, -1)
            
            print(f"   {n:2d}   |   {expected_potentials[n]:4d}   |   {actual:4d}")

verify_snn_hardware(VIVADO_LOG)
```
